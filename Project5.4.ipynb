```python
def tokenize(string, use_dictionary, length):
    def isThai(chr):
        cVal = ord(chr)
        if(cVal >= 3584 and cVal <= 3711):
            return True
        return False
    
    def search(word,source):
        places = []
        len_word = len(word)
        len_source = len(source)
        first_char = word[0]
        for x in range(0, len_source):
            if source[x] == first_char:
                if source[x:x + len_word] == word:
                    places.append(x)
        return places
    
    string = ''.join(list(filter(isThai, string)))
    string = string.replace(' ','')
    result = dict() 
    for word in use_dictionary:
        if (len(word) > length):
            s = search(word,string)
            for i in s:
                result[i] = word
            string = string.replace(word,'฿'*len(word))  
    return [value for (key, value) in sorted(result.items())]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
cv = TfidfVectorizer(tokenizer=lambda x:x.split(),ngram_range = (1,1), stop_words =["ฉัน","ผม"],
                     min_df=10,max_df=0.9,norm=None,
                     use_idf=False,smooth_idf=False)
cv.fit(train_clean)
X = cv.transform(train_clean).toarray()
y = train_label_number
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range=(0, 1))
scaler.fit(X_train)
X_train_norm =  scaler.transform(X_train)
X_test_norm = scaler.transform(X_test)
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import cross_val_score
clf = MultinomialNB()
scores = cross_val_score(clf,X_train,y_train, cv=10)
scores.mean()
```

```python
clf = MultinomialNB()
clf.fit(X_train,y_train)
```

```python
y_predict = clf.predict(X_test)
print("Test accuracy : "+str(sum(y_test == y_predict)/len(y_test)*100))
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
clf = LogisticRegression(random_state=0)
np.mean(cross_val_score(clf, X_train_norm, y_train, cv=10))
```
